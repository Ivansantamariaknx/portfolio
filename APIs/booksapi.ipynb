```python
import requests
import pandas as pd
```

```python
url="https://openlibrary.org/subjects/fantasy.json&limit=50"
#subject must be changed for a genre (e.g  fantasy)
```

```python
response=requests.get(url)
response
```

```python
data=response.json()
data
```

## Getting a dataset

```python
df=pd.json_normalize(data["works"],sep="_")
df
```

```python
df.columns
```

```python
interesting_columns=['title',"authors",'subject','first_publish_year']
df=df[interesting_columns]
```

```python
df
```

```python
df['authors'] = df['authors'].apply(lambda x: ', '.join(a['name'] for a in x) if isinstance(x, list) else '')
df['subject'] = df['subject'].apply(lambda x: ', '.join(x[:5]) if isinstance(x, list) else '')  # Show up to 5 subjects
```
