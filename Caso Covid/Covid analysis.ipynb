## Find the highest contagion in one day per country

```python
country_aggregation.loc["China"].diff().max()
```

```python
country_aggregation.loc["Italy"].diff().max()
```

```python
country_aggregation.loc["Spain"].diff().max()
```

```python
#highest infection for all countries

countries=list(country_aggregation.index)
max_infection_rates= []
for c in countries:
    max_infection_rates.append(country_aggregation.loc[c].diff().max())

country_aggregation["max_infection_rate"]= max_infection_rates
```
