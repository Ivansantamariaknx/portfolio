```python
#al estar en object hay que convertir los datos a float, hay miles por tanto hay que quitar el ".", una vez quitemos esto cambiamos el decimal de . por , y pasamos astypefloat
```

```python
# Convertir comas a puntos y luego a números decimales
# Reemplazar comas con puntos
# Eliminar puntos (tratando como separador de miles)
dfed['Total'] = dfed['Total'].str.replace('.', '')

# Reemplazar comas con puntos para decimales



```

```python
dfed['Total'] = dfed['Total'].str.replace(',', '.').astype(float)
```

```python
dfed.info()
```

```python
dfed_pivot = dfed.pivot_table(index='Comunidad', columns=['Formacion', 'Periodo'], values='Total')
print(dfed_pivot.head())
```

```python
#Los datos del otro dataframe son anuales y en este caso trimestrales, es por tanto necesario obtener un valor anual:

# 1. Crear la columna 'Año'
#dfed["Año"]=dfed["Periodo"]

dfed['Año'] = dfed['Periodo'].str.extract('(\d{4})') # Extrae los 4 dígitos del año

# 2. Agrupa y calcula la media
dfed_media = dfed.groupby(['Comunidad', 'Formacion', 'Año'])['Total'].mean().reset_index()

print(dfed_media.head())
```
